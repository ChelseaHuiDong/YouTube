```python
#change trending_date, publish_time into datetime object
#read from json to get categories
us_video = pd.read_csv("USvideos.csv", parse_dates = [1, 5])
#us_video.head()
#us_video['trending_date'].head()
#trending date still not parsed try manual convert
us_video['trending_date'] = pd.to_datetime(us_video['trending_date'], format = '%y.%d.%m')
us_video.head()
```

```python
with open('us_category_id.json') as json_data:
    d = json.load(json_data)
```

```python
#extract category type
#print(d['items'][5]['snippet']['title'])
cat_type = list()
cat_id = list()
for i in range(len(d['items']) - 1):
    cat_type.append(d['items'][i]['snippet']['title'])
for i in range(len(d['items']) - 1):
    cat_id.append(int((d['items'][i]['id'])))
```

```python
cat_dict = {'cat_type':cat_type, 'category_id':cat_id}
cat_frame = pd.DataFrame(data = cat_dict)
```

```python
cat_frame.head()
```

```python
#join two frames 
us_video_init = us_video.join(cat_frame.set_index('category_id'), on = 'category_id')
us_video_init.head()
```
